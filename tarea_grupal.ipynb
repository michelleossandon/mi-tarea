```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score






datos=pd.read_csv("C:/Users/eilee/OneDrive/Escritorio/Michelle/archive/exams.csv", header= None)

print(datos.head())
```

Primero importamos la libreria e impotamos nuestro csv, el nuestros es sobre una muestra de estudiantes que dan tres examenes, matematicas, comprención lectora y el de escritura

```python
raw_data = pd.read_csv("C:/Users/eilee/OneDrive/Escritorio/Michelle/archive/exams.csv")
```

```python
plt.scatter(x=raw_data["math score"], y=raw_data["reading score"])
```

Luego hacemos un gráfico de dispersión

```python
raw_data = raw_data.sample(frac=1)

test_data = raw_data.iloc[24:]
train_data = raw_data.iloc[:24]

x_train = train_data["math score"].values.reshape(-1,1)
y_train = train_data["reading score"]

x_test = test_data["math score"].values.reshape(-1,1)
y_test = test_data["reading score"]
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)

print("accuracy: ",r2_score(y_test,y_pred)*100,"%")
```

Con las tres anteriores entrenamos nuestro modelo y vimos que nivel de cooncordania tendra, no es muy alto pero repitiendo el proceso es el máximo que alcanzamos 

```python
plt.plot(list(range(1,100)),[model.coef_[0] * x +model.intercept_ for x in range(1,100) ], color="red")
plt.scatter(x=x_test, y=y_test, color = "blue")
plt.scatter(x = x_train,y= y_train,color="green")

plt.xlabel('Math score')
plt.ylabel('Reading score')


plt.show()
```

imprimimos nuestra regresión lineal, el grafico de dispersión con la linea de tendencia, creo que así se llama

```python
import numpy as np
import matplotlib.pyplot as plt

# Calcular la correlación entre las variables "Math score" y "Reading score"
corr = np.corrcoef(x_train.reshape(-1), y_train)[0, 1]

# Graficar la regresión lineal y los puntos de datos
plt.plot(list(range(1, 100)), [model.coef_[0] * x + model.intercept_ for x in range(1, 100)], color="red")
plt.scatter(x=x_test, y=y_test, color="blue")
plt.scatter(x=x_train, y=y_train, color="green")

# Etiquetas de los ejes y título del gráfico
plt.xlabel('Math score')
plt.ylabel('Reading score')

# Calcular el intercepto y la pendiente de la regresión lineal
intercept = model.intercept_
slope = model.coef_[0]

# Imprimir la correlación, el intercepto y la función de la regresión lineal
print(f"Correlación: {corr:.2f}")
print(f"Intercepto: {intercept:.2f}")
print(f"Función de la regresión lineal: y = {slope:.2f}x + {intercept:.2f}")

plt.show()
```

imprimimos las resultaods, correlación, la intercepción con el eje Y y la funsión f(x), tambien le pusimos nombre a los ejes, y colores a los elementos, al igual que arriba, estos fueron nuestros resultados
